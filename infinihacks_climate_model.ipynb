```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
  tf.keras.layers.InputLayer(input_shape=[2]),
  tf.keras.layers.Dense(20, name="hidden_layer_0"),
  tf.keras.layers.Dense(12, name="hidden_layer_1"),
  # tf.keras.layers.Dense(12, name="hidden_layer_2"),
  tf.keras.layers.Dense(len(y_train.columns), name="output_layer")                                       
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["mae"])

from tensorflow.keras.utils import plot_model
model.summary() 
plot_model(model, show_shapes=True, show_dtype=True, show_layer_names=True)
```

```python
model_history = model.fit(X_train, y_train, 
                          epochs=10000,
                          callbacks=[tf.keras.callbacks.EarlyStopping(monitor='loss', patience=10)],
                          verbose=1, 
                          batch_size = 16)
pd.DataFrame(model_history.history).plot(kind ='line')
```

```python
eval_history = model.evaluate(X_test, y_test)
```
